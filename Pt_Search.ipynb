# Klassen f√ºr unterschiedliche Nuggettypen

```python
class Nugget():
    def __init__(self, titel):
        self.name = 'Nugget'
        self.titel = titel
        self.ID = 0
        
        
class MotNugget(Nugget):
    def __init__(self, titel):
        super().__init__(titel)
        self.color = 'orange'
        self.name = 'Motivation'
class InhNugget(Nugget):
    def __init__(self, titel):
        super().__init__(titel)
        self.color = 'green'
        self.name = 'Inhalt'
class BspNugget(Nugget):
    def __init__(self, titel):
        super().__init__(titel)
        self.color = 'red'
        self.name = 'Beispiel'
class PruefNugget(Nugget):
    def __init__(self, titel):
        super().__init__(titel)
        self.color = 'blue'
        self.name = 'Pr√ºfung'

        
```

# Klasse f√ºr die Kontrollstruktur 

```python
class  ControlStructureNugget(Nugget):
    def __init__(self, titel):
        super().__init__(titel)
        self.name = 'Kontrollstruktur'
        self.color = 'gray'
        self.nuggets: dict[Nugget] = []
        self.titel = titel
    def addNugget(self,nugget:Nugget): 
        
        if self.validate(nugget, self.nuggets):
            print("YES!")
            self.nuggets.append(nugget)
        else:
            print("NO!")
        
    def delNug(self):
        del self.nuggets[-1]
        print(self.nuggets)

    def validate(self,nugget, nuggetlist):
        locBool = False
# Regel "Als ersts Bedarf es einer Motivation"
        if not nuggetlist:
            print('not nug')
            if type(nugget) is MotNugget:
                print("Ein Modul muss mit einem Motivationsnugget beginnen")
                locBool = True
        if nuggetlist:
            print('nug')
    
            
# Regeln f√ºr InhaltsNuggets            
            if type(nugget) is InhNugget:
                locBool = False
                print(type(nuggetlist[-1]))
# Regel "Ein InhaltsNugget muss einer Motivation folgen        
                if type(nuggetlist[-1]) is MotNugget:
                    locBool = True
                if type(nuggetlist[-1]) is BspNugget:
                    locBool = True
                
                
# Regeln f√ºr MotivationsNuggets                            
            if type(nugget) is MotNugget:
                locBool = False                
# Regel "Ein MotivationsNuggets kann einem Pr√ºfungsnugget folgen        
                if type(nuggetlist[-1]) is PruefNugget:
                   locBool = True
            
# Regel "Ein MotivationsNuggets kann einem Kontrollstruktursnugget folgen        
                if type(nuggetlist[-1]) is ControlStructureNugget:
                   locBool = True                                  
         
                if type(nuggetlist[-1]) is AlternateControlStructureNugget:
                   locBool = True  
        
        
# Regeln f√ºr BeispielNuggets                            
            if type(nugget) is BspNugget:
                locBool = False
# Regel "Ein BeispielNugget muss einem Inhaltsnugget folgen        
                if type(nuggetlist[-1]) is InhNugget:
                    locBool = True
                
                
                
# Regeln f√ºr Pr√ºfNuggets                            
            if type(nugget) is PruefNugget:
                locBool = False
# Regel "Ein Pr√ºfNugget muss einem Bsp folgen        
                if type(nuggetlist[-1]) is BspNugget: 
                    locBool = True

            
# Regeln f√ºr KontrollStrukNugget                            
            if type(nugget) is ControlStructureNugget:
                locBool = False
                print(type(self.nuggets[-1]))
# Regel "Ein KontrollStrukturNugget kann einem Motivationsnugget folgen        
                if type(nuggetlist[-1]) is MotNugget: 
                    locBool = True
# Regel "Ein KontrollStrukturNugget kann einem Pr√ºfungsnugget folgen            
                if type(nuggetlist[-1]) is PruefNugget: 
                    locBool = True        
            
            
            
            if type(nugget) is AlternateControlStructureNugget:    
                if type(nuggetlist[-1]) is MotNugget:
                    locBool = True     
        return locBool   
    
    
class AlternateControlStructureNugget(ControlStructureNugget):
    def __init__(self, titel):
        super().__init__(titel)
        self.color = 'green'
        self.name = 'AlternateControl'
        self.alternatives = 2
          
```

# Klasse zum Ausgeben der Kursstruktur im Notebook

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Wedge


class Myprinter():

    
    def print(self,nuggets):
    #define Matplotlib figure and axis
        fig, ax = plt.subplots()

    #create simple line plot
        ax.plot([0, 0],[0, 0])
        i = 0
        for x in nuggets:
            print(type(x))
            if type(x) is MotNugget :
                ax.add_patch(Rectangle((1*i, 0), 1, 1, color=x.color))
            if type(x) is InhNugget :
                ax.add_patch(Rectangle((1*i, 0), 1, 1, color=x.color))
            if type(x) is BspNugget :
                ax.add_patch(Rectangle((1*i, 0), 1, 1, color=x.color))
            if type(x) is PruefNugget :
                ax.add_patch(Rectangle((1*i, 0), 1, 1, color=x.color)) 
            if type(x) is AlternateControlStructureNugget :
                i=i+2
                u=i
                y=0    
                loc = 0
                while loc < x.alternatives:
                    i=u
                    i=i-1
                    
                    theta1, theta2 = 90, 90 + 180
                    radius = 0.7
                    center = (i, 0.5+loc*2)
                    w1 = Wedge(center, radius, theta1, theta2, fc='gray', edgecolor='black')
                    i=i+0.5
                    for nug in x.nuggets:
                        print('Nug')
                        if type(nug) is MotNugget :
                            ax.add_patch(Rectangle((1*i, loc*2), 1, 1, color=nug.color))
                        if type(nug) is InhNugget :
                            ax.add_patch(Rectangle((1*i, loc*2), 1, 1, color=nug.color))
                        if type(nug) is BspNugget :
                            ax.add_patch(Rectangle((1*i, loc*2), 1, 1, color=nug.color))
                        if type(nug) is PruefNugget :
                            ax.add_patch(Rectangle((1*i, loc*2), 1, 1, color=nug.color))    
                        i=i+1.5
                    theta3, theta4 = 270,450
                    i=i+1
                    center = (i, 0.5+loc*2)
                    
                    w2 = Wedge(center, radius, theta3, theta4, fc='gray', edgecolor='black')
                    ax.add_patch(w1)
                    
                    ax.add_artist(w2)
                    loc=loc+1
               
                y=y+1
            if type(x) is ControlStructureNugget :   
                i=i+1
                theta1, theta2 = 90, 90 + 180
                radius = 0.7
                center = (i, 0.5)
                w1 = Wedge(center, radius, theta1, theta2, fc='black', edgecolor='black')
                i=i+0.5
                for nug in x.nuggets:
                    print('Nug')
                    if type(nug) is MotNugget :
                        ax.add_patch(Rectangle((1*i, 0), 1, 1, color=nug.color))
                    if type(nug) is InhNugget :
                        ax.add_patch(Rectangle((1*i, 0), 1, 1, color=nug.color))
                    if type(nug) is BspNugget :
                        ax.add_patch(Rectangle((1*i, 0), 1, 1, color=nug.color))
                    if type(nug) is PruefNugget :
                        ax.add_patch(Rectangle((1*i, 0), 1, 1, color=nug.color))    
                    i=i+1.5
                theta3, theta4 = 270,450
                i=i+1
                center = (i, 0.5)
        
                w2 = Wedge(center, radius, theta3, theta4, fc='black', edgecolor='black')
                ax.add_patch(w1)

                ax.add_artist(w2)

      


            i=i+0.2
            i=i+1

#display plot
        plt.show()
```

```python
from flask import Flask, json, request, jsonify


api = Flask(__name__)

@api.route('/mainControl', methods=['GET'])
def get_mainControll():
    return json.dumps(m, default=vars)

# GET requests will be blocked
@api.route('/addControl', methods=['POST'])
def addControl():
    request_data = request.get_json()

    name = None
    titel = None
    python_version = None
    example = None
    boolean_test = None

    if request_data:
        if 'name' in request_data:
            titel = request_data['titel']
            name = request_data['name']
            if name == 'Motivation':
                motNug = MotNugget(titel)
                m.addNugget(motNug)
            if name == 'Inhalt':
                inhNug= InhNugget(titel)
                m.addNugget(inhNug)
            if name == 'Beispiel':
                bspNug = BspNugget(titel)
                m.addNugget(bspNug)
            if name == 'Pr√ºfung':
                prNug = PruefNugget(titel)
                m.addNugget(prNug)



    return ' Size: ' + str(len(m.nuggets)) 
   
if __name__ == '__main__':
    api.run() 
```
